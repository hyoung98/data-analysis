```python
p_df.head()
```

```python
def parse_region(data):
    parse = data.split(" ")[0] # 뒤에 있는 문자 가져오기
    parse = parse.replace(' ', '') # 공백 없애기
    return parse
```

```python
p_df["행정구역"] = p_df["행정구역"].apply(parse_region)
```
